```
from lxml import html, etree
import pandas as pd
from urllib.request import urlopen

def page_scrape(file_name):
  with open(file_name, "r") as f:
    page = f.read()
  tree = html.fromstring(page)
  Selector_0 = "div.restaurants-list-ListCell__nameBlock--1hL7F span a"
  elements = tree.cssselect(Selector_0)
  links = []

  for element in elements:
    l = element.get('href')
    links.append(l)
    
  df = pd.DataFrame(links)

  res_name = []
  address = []
  contact_no = []
  menu_type = []
  review = []

  for i in range(len(df)):

    html_link = urlopen(links[i]).read()
    tree = html.fromstring(html_link)
    sel = "h1.ui_header"
    sel1 = "span.detail span"
    sel2 = "div.header_links a"
    sel3 = "a.ui_link span.detail"
    sel4 = "div.prw_rup.prw_common_bubble_rating.rating span.ui_bubble_rating"

    a = '{}'.format(tree.cssselect(sel)[0].text)
    b = '{} '.format(tree.cssselect(sel1)[0].text) + '{}'.format(tree.cssselect(sel1)[1].text) + '{}'.format(tree.cssselect(sel1)[2].text) 
    d = '{}'.format(tree.cssselect(sel3)[0].text)

#  ff = '{}'.format(tree.cssselect(sel5)[0].text)
# c = '{}'.format(tree.cssselect(sel2)[1].text)

    c = tree.cssselect(sel2)
    c1 = []
    for all_c in c:
      if all_c.text == 'Indienne':
        a1 = '{}'.format(all_c.text)
        c1.append(a1)
      else:
        c1 = c1
  
    e = []
    el1 = tree.cssselect(sel4)
    for el in el1:
      l1 = el.get('alt')
      e.append(l1)

    res_name.append(a)
    address.append(b)
    menu_type.append(c1)
    contact_no.append(d)
    review.append(e)

    i = i + 1
#  if i == 3: break
  
  df = pd.DataFrame({'Restaurant_name': res_name, 'Address': address, 'Contact_no': contact_no, 'Menu_type': menu_type, 'Reviews': review})
  print(df.shape)
  
  return page_scrape
 
print(page_scrape(file_name='/content/drive/My Drive/Colab Notebooks/restaurants_file/restaurants.html'))
```
